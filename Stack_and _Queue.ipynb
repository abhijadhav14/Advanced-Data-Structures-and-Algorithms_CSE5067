```python
# Stack implementation using list

stack = []  # empty stack

# Function to push an element onto the stack
def push():
    element = input("Enter element to push: ")
    stack.append(element)
    print(f"{element} pushed onto stack.")

# Function to pop an element from the stack
def pop():
    if not stack:
        print("Stack is empty! Nothing to pop.")
    else:
        element = stack.pop()
        print(f"Popped element: {element}")

# Function to peek the top element of the stack
def peek():
    if not stack:
        print("Stack is empty!")
    else:
        print(f"Top element: {stack[-1]}")

# Menu-driven program
while True:
    print("\n--- Stack Operations ---")
    print("1. Push")
    print("2. Pop")
    print("3. Peek")
    print("4. Display Stack")
    print("5. Exit")

    choice = int(input("Enter your choice (1-5): "))

    if choice == 1:
        push()
    elif choice == 2:
        pop()
    elif choice == 3:
        peek()
    elif choice == 4:
        print("Current Stack:", stack)
    elif choice == 5:
        print("Exiting program...")
        break
    else:
        print("Invalid choice! Please try again.")
```

```python
# Simple Queue Implementation

queue = []  # empty queue

# Enqueue operation
def enqueue():
    element = input("Enter element to enqueue: ")
    queue.append(element)
    print(f"{element} added to the queue.")

# Dequeue operation
def dequeue():
    if not queue:
        print("Queue is empty! Nothing to dequeue.")
    else:
        element = queue.pop(0)
        print(f"Dequeued element: {element}")

# Display front element
def peek():
    if not queue:
        print("Queue is empty!")
    else:
        print(f"Front element: {queue[0]}")

# Menu-driven program
while True:
    print("\n--- Simple Queue Operations ---")
    print("1. Enqueue")
    print("2. Dequeue")
    print("3. Peek")
    print("4. Display Queue")
    print("5. Exit")

    choice = int(input("Enter your choice (1-5): "))

    if choice == 1:
        enqueue()
    elif choice == 2:
        dequeue()
    elif choice == 3:
        peek()
    elif choice == 4:
        print("Current Queue:", queue)
    elif choice == 5:
        print("Exiting program...")
        break
    else:
        print("Invalid choice! Please try again.")
```

```python
# Circular Queue Implementation

size = int(input("Enter the size of the Circular Queue: "))
queue = [None] * size
front = -1
rear = -1

def enqueue():
    global front, rear
    element = input("Enter element to enqueue: ")
    if (rear + 1) % size == front:
        print("Queue is Full! Cannot enqueue.")
    else:
        if front == -1:
            front = 0
        rear = (rear + 1) % size
        queue[rear] = element
        print(f"{element} added to the Circular Queue.")

def dequeue():
    global front, rear
    if front == -1:
        print("Queue is Empty! Nothing to dequeue.")
    else:
        print(f"Dequeued element: {queue[front]}")
        if front == rear:
            front = rear = -1
        else:
            front = (front + 1) % size

def display():
    if front == -1:
        print("Queue is Empty!")
    else:
        print("Circular Queue Elements:", end=" ")
        i = front
        while True:
            print(queue[i], end=" ")
            if i == rear:
                break
            i = (i + 1) % size
        print()

# Menu-driven program
while True:
    print("\n--- Circular Queue Operations ---")
    print("1. Enqueue")
    print("2. Dequeue")
    print("3. Display Queue")
    print("4. Exit")

    choice = int(input("Enter your choice (1-4): "))

    if choice == 1:
        enqueue()
    elif choice == 2:
        dequeue()
    elif choice == 3:
        display()
    elif choice == 4:
        print("Exiting program...")
        break
    else:
        print("Invalid choice! Please try again.")
```

```python
# Priority Queue Implementation using List

queue = []  # Empty priority queue

# Function to add an element with a given priority
def enqueue():
    element = input("Enter element to enqueue: ")
    priority = int(input("Enter priority (lower number = higher priority): "))
    queue.append((priority, element))
    queue.sort()  # Sort based on priority (ascending)
    print(f"{element} added with priority {priority}.")

# Function to remove element with highest priority (lowest number)
def dequeue():
    if not queue:
        print("Priority Queue is empty! Nothing to dequeue.")
    else:
        element = queue.pop(0)
        print(f"Dequeued element: {element[1]} (Priority: {element[0]})")

# Function to display the current queue
def display():
    if not queue:
        print("Priority Queue is empty!")
    else:
        print("Current Priority Queue:")
        for p, e in queue:
            print(f"Element: {e}, Priority: {p}")

# Menu-driven program
while True:
    print("\n--- Priority Queue Operations ---")
    print("1. Enqueue (Add Element)")
    print("2. Dequeue (Remove Highest Priority)")
    print("3. Display Queue")
    print("4. Exit")

    choice = int(input("Enter your choice (1-4): "))

    if choice == 1:
        enqueue()
    elif choice == 2:
        dequeue()
    elif choice == 3:
        display()
    elif choice == 4:
        print("Exiting program...")
        break
    else:
        print("Invalid choice! Please try again.")
```
